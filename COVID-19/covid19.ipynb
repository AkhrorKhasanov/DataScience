```python
import pandas as pd # We import pandas
```

```python
import matplotlib.pyplot as plt # We use matplotlib
```

```python
df = pd.read_csv('archive/country_wise_latest.csv') # We read all of the data from this file
```

```python
df # We see dataframe
```

```python
df = df.rename(columns={"Country/Region": "Country"}) # We rename column from Country/Region to Country
```

```python
df
```

```python
filt = df['Country'] == 'Uzbekistan' # We filter the info
```

```python
uzb = df[filt] # We get the info of Uzbekistan
uzb
```

```python
# We get the info which we need
deaths = uzb['Deaths'].values[0]
recovered = uzb['Recovered'].values[0]
active = uzb['Active'].values[0]
confirmed = uzb['Confirmed'].values[0]
```

```python
vals = [deaths, recovered, active] # We create values list
labels = ['Deaths', 'Recovered', 'Active'] # Here are labels
```

```python
# Below is pie chart
plt.figure(figsize=(6, 6))
plt.pie(vals, labels=labels, autopct="%.2f%%", startangle=90)
plt.title(f"COVID-19 Distribution - Uzbekistan\nConfirmed: {confirmed}")
plt.show()
```

```python
top5 = df[df['Country'] != 'Uzbekistan'].nlargest(5, 'Deaths') # Top 5 countries with deaths (without Uzbekistan)
top5
```

```python
top5withUzbekistan = pd.concat([top5, uzb]) # Top 5 countries and Uzbekistan
top5withUzbekistan
```

```python
colors = ['green' if c == 'Uzbekistan' else 'skyblue' for c in top5withUzbekistan['Country']] # Colors of Countries
```

```python
# Bar chart
plt.figure(figsize=(6, 6))
plt.bar(top5withUzbekistan['Country'], top5withUzbekistan['Deaths'], color=colors)
plt.title('Top 5 Countries by Deaths + Uzbekistan')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
# barh chart
regions = df.groupby('WHO Region')['Deaths'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.barh(regions['WHO Region'], regions['Deaths'], color='cyan')
plt.title("Total Deaths by WHO Region")
plt.xlabel("Total Deaths")
plt.ylabel("WHO Region")
plt.show()
```
