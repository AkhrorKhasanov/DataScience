```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('grades.csv')
print(df)
```

```python
# O'rtacha baholar (har bir fan bo'yicha)
subject_means = df.mean(numeric_only=True)
print("\nFanlar bo'yicha o'rtacha baholar:")
print(subject_means)
```

```python
# Har bir talabaning o'rtacha bahosi
df['average'] = df[['math', 'physics', 'english']].mean(axis=1)
print("\nTalabalar bo'yicha o'rtacha baho")
print(df[['name', 'average']])
```

```python
# Numpy bilan max/min
math_scores = df['math'].to_numpy()
print("\nMatematika bo'yicha eng yuqori baho:", np.max(math_scores))
print("\nMatematika bo'yicha eng past baho:", np.min(math_scores))
```

```python
# Grafik – fanlar bo‘yicha o‘rtacha baho
plt.bar(subject_means.index, subject_means.values, color='skyblue')
plt.title("Fanlar bo'yicha o'rtacha baho")
plt.xlabel("Fanlar")
plt.ylabel("O'rtacha baho")
plt.show()
```

```python
# Grafik – talabalar o‘rtacha bahosi
plt.figure(figsize=(7, 5))
plt.bar(df['name'], df['average'], color='orange')
plt.title("Talabalar o'rtacha bahosi")
plt.xlabel('Talaba')
plt.ylabel('Baho')
plt.show()
```
